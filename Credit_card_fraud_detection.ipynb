```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
df =credit_data= pd.read_csv('/content/creditcard.csv')
```

the data is is extremely unbalanced 
0 represents normal transactions
1 represents fraudulant transactions

```python
lawful = credit_data[credit_data.Class == 0]
unlawful = credit_data[credit_data.Class == 1]
```

```python
print(lawful.shape)
print(unlawful.shape)
```

```python
#stats meseaure
lawful.Amount.describe()
```

```python
unlawful.Amount.describe()
```

```python
#comparing the values of the lawful and unlawful transactions
credit_data.groupby('Class').mean()
```

Building a sample dataset containing similar distribution of lawful and unlawful transactions


number of unlawful transactions == 385

```python
lawful_sample = lawful.sample(n=385)
```

linking 2 data frames

```python
new_dataset = pd.concat([lawful_sample,unlawful],axis=0 )
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spilting the data into features and targets here

```python
x = new_dataset.drop(columns='Class', axis = 1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

spilting the data into train and test 

```python
x_train ,x_test,y_train,y_test= train_test_split(x , y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

training the model using logistic regression 

```python
Model =LogisticRegression()
```

```python
Model.fit(x_train,y_train)
```

```python
#finding out the accuracy on trainning data 
x_train_pred = Model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_pred,y_train)
```

```python
print('accuracy on trainning data is :',training_data_accuracy )
```

```python
#now on test data 
x_test_pred = Model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred, y_test)
```

```python
print('accuracy on trainning data is :',test_data_accuracy )
```
